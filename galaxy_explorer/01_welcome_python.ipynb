## Part 2 — Maths

Python is an excellent calculator. Here are the main operators:

| Symbol | Meaning | Example |
|--------|---------|----------|
| `+` | add | `3 + 2` → `5` |
| `-` | subtract | `10 - 4` → `6` |
| `*` | multiply | `3 * 4` → `12` |
| `/` | divide | `10 / 4` → `2.5` |
| `**` | power | `2 ** 8` → `256` |

Astronomers deal with *enormous* numbers. Python handles them easily.

```python
# How long does light take to reach us from the Sun?

distance_to_sun_km = 149_600_000    # km (the _ just makes it easier to read)
speed_of_light_km_s = 299_792       # kilometres per second

time_seconds = distance_to_sun_km / speed_of_light_km_s
time_minutes = time_seconds / 60

# The f"..." syntax lets us insert variable values directly into text
print(f"The Sun is {distance_to_sun_km:,} km away.")
print(f"Light takes {time_seconds:.0f} seconds = {time_minutes:.1f} minutes to reach us.")
print()
print("This means: when you look at the Sun, you are seeing it as it was 8 minutes ago!")
```

```python
# How far away is the nearest star to our Sun?
# Proxima Centauri is 4.24 light years away.
# One light year = distance light travels in one year.

seconds_per_year = 365.25 * 24 * 3600       # seconds in a year
light_year_km    = speed_of_light_km_s * seconds_per_year

proxima_distance_km = 4.24 * light_year_km

print(f"One light year = {light_year_km:.2e} km")
print(f"Proxima Centauri is {proxima_distance_km:.2e} km away")
print()
print("The '.2e' means: show 2 decimal places in scientific notation (e.g. 4.01e+13).")
print("We need scientific notation because the numbers are TOO BIG to write normally!")
```

**Challenge:** How long (in years) would it take a spacecraft travelling at 50,000 km/h to reach Proxima Centauri? The fastest spacecraft ever built was the Parker Solar Probe at about 692,000 km/h.

```python
# Your turn! Fill in the blanks and run the cell.

spacecraft_speed_kmh = 692_000     # km/h (Parker Solar Probe)
spacecraft_speed_kms = spacecraft_speed_kmh / 3600  # convert to km/s

travel_time_seconds = proxima_distance_km / spacecraft_speed_kms
travel_time_years   = travel_time_seconds / seconds_per_year

print(f"At {spacecraft_speed_kmh:,} km/h, it would take {travel_time_years:,.0f} years.")
print("Humans have only been a species for about 300,000 years!")
```

## Part 3 — Lists

A **list** is a collection of values stored in one variable.
You make a list using square brackets `[  ]` with commas between items.

```python
# A list of planets in our Solar System
planets = ["Mercury", "Venus", "Earth", "Mars",
           "Jupiter", "Saturn", "Uranus", "Neptune"]

print("Planets:", planets)
print("Number of planets:", len(planets))   # len() counts items in a list
print()

# Access individual items using square brackets and an index number
# IMPORTANT: Python starts counting from 0, not 1 !
print("First planet (index 0):",  planets[0])
print("Third planet (index 2):",  planets[2])
print("Last planet  (index -1):", planets[-1])   # -1 means 'last'
```

```python
# Lists can also hold numbers
# These are the average distances from the Sun in Astronomical Units (AU)
# 1 AU = Earth-Sun distance

distances_au = [0.39, 0.72, 1.00, 1.52, 5.20, 9.58, 19.2, 30.1]

# How far is Jupiter compared to Earth?
jupiter_distance_au = distances_au[4]   # index 4 = Jupiter
earth_distance_au   = distances_au[2]   # index 2 = Earth

print(f"Jupiter is {jupiter_distance_au / earth_distance_au:.1f}x further from the Sun than Earth.")
```

---

## Part 4 — Loops

A **loop** repeats an action for every item in a list.
This is one of the most useful things in all of programming — imagine doing the same calculation for millions of galaxies!

```python
# Print each planet and its distance
# 'for ... in ...' goes through the list one item at a time

print("Planet distances from the Sun:")
print("-" * 40)   # print a line of dashes

for i in range(len(planets)):
    planet   = planets[i]
    distance = distances_au[i]
    print(f"  {planet:<10}  {distance:>5} AU")  # neat formatting
```

```python
# A neater way to loop over two lists at once: zip()
# zip() pairs up items from two lists

print("Planet distances in km:")
print("-" * 50)

for planet, dist_au in zip(planets, distances_au):
    dist_km = dist_au * 149_600_000
    print(f"  {planet:<10} = {dist_km:>18,.0f} km")
```

## Part 5 — Functions

A **function** is a reusable block of code. You write it once, then call it with different inputs.

Functions are essential for science — you write a calculation once and apply it to millions of objects.

```python
# Define a function with 'def', give it a name, list its inputs in ()
# The 'return' line sends a value back to whoever called the function

def au_to_km(au):
    """Convert a distance in Astronomical Units to kilometres."""
    km_per_au = 149_600_000
    return au * km_per_au


# Use (call) the function
earth_km   = au_to_km(1.00)
jupiter_km = au_to_km(5.20)
neptune_km = au_to_km(30.1)

print(f"Earth:   {earth_km:>20,.0f} km from the Sun")
print(f"Jupiter: {jupiter_km:>20,.0f} km from the Sun")
print(f"Neptune: {neptune_km:>20,.0f} km from the Sun")
```

```python
# A more useful function: calculate light travel time

def light_travel_time_minutes(distance_km):
    """How many minutes does light take to travel this distance?"""
    speed_of_light = 299_792   # km/s
    time_seconds   = distance_km / speed_of_light
    time_minutes   = time_seconds / 60
    return time_minutes


print("Light travel times from the Sun:")
for planet, dist_au in zip(planets, distances_au):
    dist_km = au_to_km(dist_au)                         # call function 1
    t_min   = light_travel_time_minutes(dist_km)        # call function 2
    print(f"  {planet:<10}: {t_min:>7.1f} minutes")
```

## Part 6 — Your First Plot

Now for something visual. Python has a library called **matplotlib** for making plots.

A **library** (also called a **module** or **package**) is a collection of extra tools.
You load a library with `import`.

```python
import matplotlib.pyplot as plt   # the plotting library; 'plt' is the short nickname
import numpy as np                 # numpy: fast maths on big arrays of numbers

print("Libraries loaded!")
```

```python
# Let's plot the Solar System!
# We'll draw each planet as a coloured dot at its correct distance.

planet_colours = ["gray", "wheat", "royalblue", "tomato",
                  "sandybrown", "gold", "lightblue", "steelblue"]
planet_sizes   = [4,  6,  6,  4,  20, 18, 12, 11]   # relative to real sizes

fig, ax = plt.subplots(figsize=(12, 4))

# Draw the Sun
ax.plot(0, 0, 'o', color='yellow', markersize=35, zorder=5,
        markeredgecolor='orange', markeredgewidth=2)
ax.text(0, 0.25, 'Sun', ha='center', va='bottom', fontsize=9)

# Draw each planet
for planet, dist, colour, size in zip(planets, distances_au, planet_colours, planet_sizes):
    ax.plot(dist, 0, 'o', color=colour, markersize=size, zorder=4)
    ax.text(dist, -0.25, planet, ha='center', va='top', fontsize=8, rotation=45)

# Formatting
ax.set_xlim(-1, 32)
ax.set_ylim(-0.8, 0.8)
ax.set_xlabel("Distance from Sun (AU)", fontsize=12)
ax.set_title("The Solar System (not to scale!)", fontsize=13)
ax.set_yticks([])
ax.grid(True, alpha=0.3, axis='x')

plt.tight_layout()
plt.show()
```

```python
# Let's make a more scientific plot: light travel time vs distance

distances_km    = [au_to_km(d) for d in distances_au]   # list comprehension!
times_minutes   = [light_travel_time_minutes(d) for d in distances_km]

fig, ax = plt.subplots(figsize=(9, 5))

for i, (planet, dist, t) in enumerate(zip(planets, distances_au, times_minutes)):
    ax.scatter(dist, t, s=80, color=planet_colours[i], zorder=5, edgecolors='k', linewidths=0.5)
    ax.annotate(planet, (dist, t), textcoords="offset points", xytext=(5, 5), fontsize=8)

ax.set_xlabel("Distance from Sun (AU)", fontsize=12)
ax.set_ylabel("Light travel time (minutes)", fontsize=12)
ax.set_title("How long does light take to reach each planet?", fontsize=13)
ax.grid(True, alpha=0.4)

plt.tight_layout()
plt.show()

print(f"\nFun fact: signals from the Voyager 1 probe (in interstellar space)")
print(f"take about {light_travel_time_minutes(au_to_km(160)):.0f} hours = "
      f"{light_travel_time_minutes(au_to_km(160))/60:.1f} hours to reach Earth!")
```

## Bonus: NumPy Arrays

Before we finish, let's meet **NumPy** — the most important tool for scientific Python.
Instead of ordinary Python lists, NumPy gives us **arrays**: lists of numbers that support fast maths on all elements at once.

```python
import numpy as np

# Convert our distances list to a NumPy array
distances = np.array(distances_au)

# Now we can do maths on the WHOLE array at once — no loop needed!
distances_km_array = distances * 149_600_000

print("Distances in km:", distances_km_array)
print()
print("Mean distance from the Sun:", np.mean(distances_km_array), "km")
print("Closest planet:", planets[np.argmin(distances_km_array)], "-", distances_km_array.min(), "km")
print("Furthest planet:", planets[np.argmax(distances_km_array)], "-", distances_km_array.max(), "km")
```

```python
# NumPy can also generate arrays for us
# For example: 1000 evenly spaced numbers between 0 and 2π

theta = np.linspace(0, 2 * np.pi, 1000)   # 1000 points from 0 to 2π
sine  = np.sin(theta)                       # compute sine of every point at once!

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(theta, sine, color='steelblue', linewidth=2)
ax.set_xlabel("Angle (radians)", fontsize=12)
ax.set_ylabel("sin(angle)", fontsize=12)
ax.set_title("A sine wave — the same shape as a light wave!", fontsize=13)
ax.axhline(0, color='k', linewidth=0.8)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
